```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
class testseaborn():
    def pyplottest(self):
        x=np.array([0,2,5,9,11,15,20,25,30,45,60,65,70,80,90,110])
        y=np.array([50,100,150,200,250,300,350,400,450,500,550,600,750,700,750,800])
        x1=np.array([0,2,5,9,11,15,20,25,30,45,60,65,70,80,90,110,115,120,125,130,140,150])
        y1=np.array([50,100,150,200,250,300,350,400,450,500,550,600,750,700,750,800,850,900,1000,1100,1150,1250])
        #print(x)
        #print(y)
        plt.plot(x,y,x1,y1,marker='o',color='b',linewidth=4,markersize=15,mec='g',mfc='r',ls=':')
        plt.grid(color='b',axis='x',ls='--')
        font={"family":"serif","color":'r',"size":20,}
        plt.xlabel("To the Right Side",font)
        plt.ylabel("To the upword Side",font)
        plt.title("PyPlot Test Graph",font)
        plt.figure(figsize=(20,30))
        plt.show()
ob=testseaborn()
ob.pyplottest()
```
