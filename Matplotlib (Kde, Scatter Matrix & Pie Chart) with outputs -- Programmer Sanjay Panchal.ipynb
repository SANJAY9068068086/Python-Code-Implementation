# Visualization

### Matplotlib (Kde, Scatter Matrix & Pie Chart)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.plotting import scatter_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
iris=sns.load_dataset('iris')
df=iris.drop(['species'], axis=1)
```

```python
df
```

```python
scatter_matrix(df,figsize=(8,8), diagonal='kde', color='r')  # kde Kernal Density Estimation
plt.show()                                                   #kde is just type of graph
```

```python
scatter_matrix(df,figsize=(8,8))        #if you don't choose the 'kde' it will show the bar graph bydefault
plt.show()                              #and the color is also bule bydefault
                                        #kde is just type of graph
```

```python
df.head()
```

```python
d=df.iloc[0]        #iloc means index location and [0] is index, which means only 1st row to be selected
d
```

```python
d.plot.pie(figsize=(6,6))    #pie chart and figsize is optional
```

```python
d=df.head(3).T            #.T meas transpose which means your all data change vertical to horizontal and vice-versa
d
```

```python
d.plot.pie(subplots=True, figsize=(25,30))   

#subplots means multiple graphs and also figsize is optional
#if you want to use the actual multiple chart then you have to transpose the data by .T (otherwise optional)
#the benefit of the multiple chart is that you find which portion is related to which column
#you choose head(3) then you get 3 charts
```

```python
d.plot.pie(subplots=True, figsize=(25,30), fontsize=20, autopct='%.1f')

#in this you can track the percentage cover by each.
#fontsize=size of the percentage numbers & autopct=means % that to show how much cover by each
#'%.if'= in this '%' is related to all data & '1f' is related to decimal points that you want which means 1 is for 1 decimal
#you can change the decimal points by increasing & decreasing the numbers of 'f'
```

```python
df.plot(kind='area')      #the area graph
                          #and in this you can't go in the deep to understand the graphical data
```

```python
df.plot.scatter(x='sepal_length', y='petal_length')  

#the most important ploting technique is the scatter plot
#90% we use this technique to visualization the data
#in this you have to remember that you need to have 2 sets of data always, 1st is on x-asix and 2nd is on y-axis
#we use this to find the relationship between x and y
```

```python
graph_1=df.plot.scatter(x='sepal_length', y='petal_length', label='Length') 
df.plot.scatter(x='sepal_width', y='petal_width', label='Width', ax=graph_1, color='r')

#in this we can plot one graph into another graph
#lable means to show the lables above the graph
#ax=to connect the another graph (this is the parameter to connect the another graph into this graph)
#we use red and blue color to understand this, in 1st graph we did't use the color option because bydefault is blue color
#blue color is for length graph and red color is for width graph
```
