### 1. Seaborn (Numerical Data Ploting)

#### relplot() , scatterplot() , lineplot()

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.tail()

#first of all we need to import  a dataset whatever you want
#in this we imported 'tips' dataset which available in seaborn just like 'iris' that we imported in the previous chapter
#tips is nothing but it is generally us for giving tip to someone (just like hotels and restaurents)
#this is available in the seaborn so we just imported to use is
#tail() is used to see the last 5 rows, just like head()
#this data is from USA
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)

#relplot is a part of seaborn
#relplot is a relational ploting between two columns
#all data coming from tips dataset
```

```python
tips['smoker'].value_counts()

#in  this we can select a particular column and count the number of characters inside this
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='time')

#hue and style is the parameters to understand the data easily
#see the last one, in that it is cross means Dinner and blue means Smoker and pay tips according to total bill
#hue makes the colors & style give the style
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, size='size')

#in this we use the size parameter means to understand the numbers according to given size
#here we use the 'size' option in the size parameter to understand the numbers of peoples
#you can also choose any column in the size parameter
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, size='size')

#the scatterplot is just like relplot and there is only one difference that is design
```

```python
sns.lineplot(x='total_bill', y='tip', data=tips)

#the lineplot is to define the data in the line format 
```

### 2. Seaborn (Categorical Data Ploting)

#### catplot() , boxplot() , stripplot() , swarmplot() 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(x='day', y='total_bill', data=tips)

#catplot is simple categorical variables, it is use to understand the particular variables
#her is use the day wise and you can see what happening, it shows the data according to week days
```

```python
sns.catplot(x='smoker', y='total_bill', data=tips)

#here we can easily understand by smoker data
```

```python
sns.catplot(x='smoker', y='tip', data=tips)

#here is one more example to understand this, in this now we can see the smoker wise tips
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)

#there is no difference but only design is different to understand the data
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='size')

#here is the combination with swarm , is not nothing but to show the kind parameter
#you can also give the size by using hue parameter
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)

#same like catplot(), only the one difference is diagram
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)

#we already discussed the boxplot() in the previous chapter, same understanding
```

### 3. Seaborn (Distribution Of The Data)

#### distplot() , kdeplot(), jointplot()

```python
import numpy as np
from numpy.random import randn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
x=randn(1000)  #we discussed about randn in the numpy chapter
```

```python
sns.distplot(x, kde=False, hist=True)

#we aleady discussed about kde and hist
#kde=type of graph
#hist=histogram (we already disceussed about it in the previous chapter)

#if we use the kde=True and hist=True , both the diagram will show
#if we use the kde=False and hist=True , only hist diagram will show
#if we use the kde=True and hist=False , only kde diagram will show
#if we use the kde=False and hist=False , no any disgram will show 

#This is called the distplot() means distribution ploting

#see the next 3 examples of these of the distplot()
```

```python
sns.distplot(x, kde=True, hist=False)

#Here kde is True and hist is False
```

```python
sns.distplot(x, kde=True, hist=True)

#Here is both kde and hist are True
```

```python
sns.distplot(x, kde=False, hist=False)

#Here is both kde and hist are False
```

```python
sns.kdeplot(x)

#This is the kdeplot(), we can directly use the kde plot by using this
```

```python
sns.kdeplot(x, shade=True)

#We can also give it the shade True, but by default the shade is False
```
